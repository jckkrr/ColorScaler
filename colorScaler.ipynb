### COLOR SCALER

A script to map out a series of numbers to a color gradient.

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt

import random     
```

```python
# the main function. It's current settings give a highlight from dull to bright yellow.

def colorScaler(n): 
    
    R = 1 - ((n-1)**2) / 50
    G = 1 - ((n-1)**2) / 450
    B = 0.625 - (n/1.75)
    
    if n < 0.8:
        R = 0.9 +  (n/8)
        G = 0.8 +  (n/4)
        
    else:
        R = 1
        G = 1
        
    B = 0.625 - (n/1.75)

    c = R,G,B
    
    return c
```

```python
# a function to show you the color gradient and plot the rgb lines

def graphColors(calledFunc):

    x = np.linspace(0.000001, 1, 110)

    RGBlist = [calledFunc(n) for n in x]

    fig, ax = plt.subplots(1, 2, gridspec_kw={'width_ratios': [10,15]})

    ax[0].plot(x, RGBlist)   # maths chart

    ax[1].scatter(x, .5- (x-x), marker='|', c=RGBlist, linewidth=7, s = 50000) # color gradient
    plt.rcParams["figure.figsize"] = (20,5)
```

```python
graphColors(colorScaler)
```

```python
# a fucntion to give the text a gradient

def colorScalerForText(n): 
    
    R = 1 - (n ** .25)
    B = 1 - (n ** .25)
    G = 1 - (n ** .25)

    c = R,G,B
    return c
```

```python
graphColors(colorScalerForText)
```

```python
# an example of how it might work, using random numbers on a scale of 0-100

def randomPlot():

    plt.rcParams["figure.figsize"] = (15,7)
    fig, ax = plt.subplots()

    for x in range(0,10):
        for y in range(0,5):

            n = random.uniform(0, 1)
            c = colorScaler(n)

            ax.scatter(x, y, s=4000, marker='s', c=np.array([c]))

            string_n = str(round(n*100,2))

            textc = colorScalerForText(n)

            plt.text(x, y, string_n, weight = 'bold', size=12, ha="center", va="center", c=textc)
            
randomPlot()
```
